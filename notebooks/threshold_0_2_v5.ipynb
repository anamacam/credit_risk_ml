### El Ganador: Umbral 0.20 (Threshold_0_2_V5)
Este es el "punto de equilibrio" para un negocio de riesgo crediticio por las siguientes razones:

- **Recall del 93.3%:** Estás detectando a casi todos los clientes de alto riesgo. Solo se te escapa un 6.7%. En comparación, el umbral de 0.35 baja el recall al 83.3%, lo cual es demasiado arriesgado (se te escapan el doble de morosos).

- **F1-Score (0.489):** Es el más alto de los tres experimentos. Esto confirma que es el punto donde el equilibrio entre precisión y sensibilidad es más eficiente matemáticamente.

- **Eficiencia:** El umbral de 0.05 tiene un Recall de 1.0 (perfecto), pero su precisión es bajísima (0.3). Estarías rechazando a muchísima gente buena solo por miedo, perdiendo negocio. El de 0.20 mantiene un Recall altísimo pero con una precisión más razonable (0.33).

| Métrica   | Umbral 0.05      | 🏆 Umbral 0.20 (Ganador) | Umbral 0.35 |
| --------- | ---------------- | ------------------------ | ----------- |
| Recall    | 💯 1.00 (Máximo) | 0.933                    | 0.833       |
| F1-Score  | 0.461            | 🟢 0.489                 | 0.568*      |
| Precision | 0.300            | 🟢 0.331                 | 0.431       |
| ROC AUC   | 0.786            | 🟢 0.786                 | 0.786       |

**Nota:** Aunque el F1-Score es mayor en 0.35, en riesgo crediticio siempre preferimos sacrificar un poco de F1 para mantener el Recall por encima del 90%. El de 0.20 es el más equilibrado para producción.

**CreditRiskRF_Official_Threshold** es un modelo de clasificación

- **El objetivo** es asignar una etiqueta o categoría.

- **Salida:** Discreta (0 o 1, "Good" o "Bad").

- **Caso:** Busca predecir si un cliente pertenece a la clase "Riesgo Alto" o "Riesgo Bajo".

- **Métricas que utilizadas:** Recall, F1-Score, Precision y AUC. Estas métricas solo tienen sentido cuando se compara etiquetas reales contra predicciones categóricas.
