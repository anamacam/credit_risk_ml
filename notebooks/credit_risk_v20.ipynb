## MLflow que resume el rendimiento de un modelo

**Gráficos de Radar** (diagramas de araña) que comparan diferentes métricas de rendimiento de un modelo de aprendizaje automático. En estos gráficos, cuanto más cerca esté la mancha azul del borde exterior, mejor es el desempeño en esa métrica.

Interpretación de los dos perfiles principales:

**1. Perfil Superior:** Métricas de Clasificación Estándar

El gráfico evalúa la calidad de las predicciones del modelo:

- **training_precision_score:**  Esta es la métrica más alta. Indica que cuando el modelo predice una categoría, **casi siempre acierta** (pocos falsos positivos).
- **training_recall_score:**  Este es notablemente más bajo. Esto significa que el modelo es **"conservador"** y deja pasar algunos casos reales sin detectar (falsos negativos).
- **training_f1_score:**  Es el promedio **equilibrado** entre las dos anteriores.
- **training_roc_auc:**  Muestra una excelente capacidad del modelo para **distinguir** entre clases.
  
**2. Perfil Inferior: Estabilidad y Error**

En este gráfico añade métricas de probabilidad y pérdida:

- **training_log_loss:** Esta métrica mide la **"incertidumbre"**. En este caso, el área es amplia, lo que sugiere que el modelo no solo acierta, sino que lo hace con **niveles de confianza adecuados**.
- **training_roc_score / training_accuracy_score:** Ambos muestran un rendimiento sólido y consistente, indicando que el modelo **ha aprendido bien los patrones** de los datos de entrenamiento.
  
**Comparativa (Derecha)**

Los gráficos pequeños a la derecha parecen son versiones de validación de otros modelos. Al ser más pequeños (estar más cerca del centro), indican un rendimiento inferior en comparación con el modelo principal de la izquierda.

### 1. Métricas de Rendimiento (Metrics)

Las cifras registradas indican qué tan bien está funcionando el modelo sobre los datos de entrenamiento:

- **Precision (0.8414):** De todas las veces que el modelo predijo "alto riesgo", acertó el **84.1%** de las veces.
- **Recall (0.823):** El modelo fue capaz de identificar correctamente al **82.3%** de los clientes que realmente tienen riesgo.
- **F1 Score (0.8277):** Es el equilibrio entre precisión y recall. Un valor superior a **0.8** es generalmente muy bueno.
- **Accuracy (0.823):** El modelo clasifica correctamente al **82.3%** del total de los casos.
- **ROC AUC (0.9165):** Es una métrica de excelencia. Al estar cerca de **1.0**, indica que el modelo tiene una alta capacidad para distinguir entre clientes con riesgo y sin riesgo.

### 2. Matriz de Confusión

Visualiza los aciertos y errores del modelo:

- **Verdaderos Negativos (No Risk):**  **378** casos correctamente identificados como sin riesgo.
- **Verdaderos Positivos (High Risk):**  **133** casos correctamente identificados como de alto riesgo.
- **Falsos Negativos (Abajo a la izq.):** El modelo falló al no detectar a **276** personas que sí eran de alto riesgo.
- **Falsos Positivos (Arriba a la der.):** El modelo marcó erróneamente a **79** personas como "riesgo" cuando no lo eran.

## 3. Curva ROC (ROC Curve)

El gráfico azul en la esquina inferior derecha muestra la relación entre la sensibilidad y la especificidad. Cuanto más se aleje la curva de la línea punteada diagonal y se acerque a la esquina superior izquierda, **mejor es el modelo**. Con un **AUC de 0.9165**, esto nos dice que este modelo es altamente confiable para este conjunto de datos.

### Conclusión:
El modelo de la izquierda **es muy preciso**, pero tiene un punto **débil en el recall** (sensibilidad). Dependiendo del ROI del negocio, podrías necesitar ajustes para que no se le **"escapen"** tantos casos positivos.

![texto](img\recall.png)



En un negocio de créditos bancarios, **el recall (o sensibilidad) es probablemente la métrica más crítica de todas**, porque está directamente ligada a las pérdidas financieras por impagos.

Explicación de cómo afecta un recall bajo o alto en este contexto:

**1. ¿Qué significa el Recall en créditos?** 

El recall mide la capacidad del modelo para identificar correctamente a los **clientes de alto riesgo** (los que no van a pagar).
- **Recall Alto:** Detectas a casi todos los clientes morosos.
- **Recall Bajo:** Se le escapan clientes peligrosos y les terminas dando el dinero.
  
**2. El impacto de un Recall Bajo (Modelo actual)**

Emodelo tiene una **precisión alta pero un recall más bajo**. En un banco, esto provoca lo siguiente:

- **Aumento de la Cartera Vencida:** Estás aprobando créditos a personas que el modelo no marcó como riesgo, pero que en realidad sí lo son (Falsos Negativos).
- **Pérdida de Capital:** El costo de prestar dinero a alguien que no paga es mucho mayor que el beneficio de los intereses de un buen cliente.
- **Riesgo de Solvencia:** Si muchos "falsos negativos" se acumulan, la salud financiera del banco se compromete.

**3. El balance: Recall vs. Precisión**

En banca siempre hay un conflicto de intereses:

- **Si priorizas el Recall (Ser muy estricto):** Rechazará a casi todos los morosos, pero también rechazarás a muchos clientes buenos por error **(Falsos Positivos)**, lo que significa **pérdida de oportunidades de negocio** y clientes molestos.
- **Si priorizas la Precisión (Ser muy seguro):** Solo rechaza a alguien cuando estás 100% seguro de que es malo, pero deja pasar a muchos morosos "deudosos".


### Resumen para el negocio
Para un banco, normalmente es mejor tener un Recall alto aunque la Precisión baje un poco. Es preferible perder un cliente potencial (costo de oportunidad) que perder el capital completo del préstamo (pérdida directa)

### Conflicto en los Datos

Hay una observación importante: aunque el AUC es muy alto (0.91), el número de Falsos Negativos (276) es mayor que el de Verdaderos Positivos (133). Esto sugiere que:
- 1. El modelo es bueno ordenando los riesgos, pero el **umbral de decisión (threshold)** actual podría estar muy alto.
- 2. Si baja el umbral, podrías capturar más de esos 276 casos de riesgo, aunque esto aumentaría los Falsos Positivos (las 79 personas).
