## Reporte de Optimización de Umbral (Threshold Tuning)

1. **Justificación de Parámetros Técnicos**
Para este análisis, se definieron parámetros específicos que responden a la naturaleza del **Riesgo Crediticio:**

**THRESHOLDS = np.arange(0.25, 0.46, 0.01)**

- **Por qué este rango:** En lugar de usar el estándar de 0.50, exploramos valores más bajos porque buscamos que el modelo sea más "sensible" al riesgo.

- **Granularidad (0.01):** Un salto de 0.01 nos permite encontrar el punto exacto donde el beneficio de detectar un moroso extra supera el costo de rechazar a un cliente bueno.

- **Sesgo Conservador:** Al movernos entre 0.25 y 0.45, obligamos al modelo a clasificar como "Riesgo" a cualquier cliente que tenga una probabilidad de impago superior al 25%, priorizando la seguridad del capital.

**TARGET_RECALL = 0.60**
- **Definición de Negocio:** Este es nuestro "piso" de aceptación. Significa que, como mínimo, el banco exige detectar al 60% de los clientes morosos reales.

- **Protección:** Un Recall inferior al 60% dejaría pasar a demasiados clientes insolventes, lo que resultaría en pérdidas financieras inaceptables para la institución.

**2. Explicación de Métricas Utilizadas** 

El cálculo de estas tres métricas es fundamental para tomar una decisión equilibrada:

1. **Recall (Sensibilidad):**
   - **Fórmula:** $\frac{VP}{VP + FN}$
   - **Por qué:** Es la métrica más importante en este proyecto. Mide nuestra capacidad para encontrar a los "malos pagadores". Un Recall alto significa que hay pocos Falsos Negativos (morosos que el modelo dejó pasar).
  
2. **Precision (Precisión):**
   - **Fórmula:** $\frac{VP}{VP + FP}$
   - **Por qué:** Indica qué tan acertados somos cuando el modelo dice "Bad". Si es muy baja, estamos rechazando a demasiados clientes buenos (Falsos Positivos). Nos ayuda a medir el costo de oportunidad.
  
3. **F1-Score:**
   - **Fórmula:** $2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}$
   - **Por qué:** Es el promedio armónico entre las dos anteriores. Lo usamos para asegurar que, al buscar un Recall alto, no destruyamos completamente la precisión del modelo.

**3. Definición de Checkpoints de Éxito**
Los **Checkpoints** son criterios automáticos que determinan si el modelo puede avanzar a producción o debe volver a la etapa de diseño.

- **PASSED (Éxito):** Se logra si el Recall ≥ 0.60. En nuestro caso, obtuvimos 0.82, lo que valida que la estrategia de umbral en 0.27 es sumamente efectiva.

- **FAILED (Fallo):** Si ningún umbral logra el Recall mínimo del 60%. Esto dispararía una alerta para volver a la etapa de Feature Engineering o probar algoritmos distintos.

**4. Resumen de Logros para Stakeholders**

**Identificación Proactiva:** Logramos capturar al 82% de los morosos, superando el objetivo inicial del 60%.

**Umbral Óptimo:** Se determinó que 0.27 es el punto de equilibrio donde protegemos el capital sin bloquear excesivamente la operación comercial.
